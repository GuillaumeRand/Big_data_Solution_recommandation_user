```python
df_filtered_month = customer_purchases_by_month(id_client)

# Tracer un graphique des d√©penses totales par mois
plt.bar(df_filtered_month['MOIS_VENTE'], df_filtered_month['PRIX_NET'])
plt.xlabel('Mois')
plt.ylabel('D√©penses totales')
plt.title('D√©penses totales par mois pour le client {}'.format(id_client))
plt.xticks(rotation=45)
plt.show()
```

```python
df_filtered = customer_purchases_evol(id_client)
# Tracer un graphique de tendance des d√©penses totales par mois
plt.plot(df_filtered['MOIS_VENTE'], df_filtered['PRIX_NET'], '-o')
plt.xlabel('Mois')
plt.ylabel('D√©penses totales')
plt.title('Tendance des d√©penses totales pour le client {}'.format(id_client))
plt.xticks(rotation=45)
plt.show()
```

```python
df_average_by_category = customer_purchases_by_category(id_client)

# Tracer un graphique des d√©penses moyennes par cat√©gorie
df_average_by_category.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Famille')
plt.ylabel('D√©pense moyenne')
plt.title('D√©penses moyennes par famille pour le client {}'.format(id_client))
plt.show()
```
